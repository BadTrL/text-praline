TEST OF PRALINE ON A VERY ILLUSTRATED PDF
-----------------------------------------

```python
import pathlib
from textpraline import praline
from pdfminer.high_level import extract_text

pdf_path = pathlib.Path("./corpus/docu_astro.pdf")
raw = extract_text(str(pdf_path))
output_path_raw = pathlib.Path("./output/docu_astro_extracted.txt")
output_path_raw.write_text(raw, encoding="utf-8")

cleaned, report = praline(raw, debug=True)

output_path = pathlib.Path("./output/docu_astro_cleaned.txt")
output_path.write_text(cleaned, encoding="utf-8")

print("Cleaned file written to:", output_path)
print(report)
```

-> Conclusion: almost 3000 parasites characters has been deleted, almost 3%. which signify 3% non-parasite chunks
